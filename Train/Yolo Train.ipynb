```python
import os
import shutil

# Source path
source = '/kaggle/input/voc-dataset/data_images/data_images'

# Destination path
destination = '/kaggle/working/yolov5'

# Create destination directory if it doesn't exist
if not os.path.exists(destination):
    os.makedirs(destination)

# Function to move files
def move_files(src, dst):
    try:
        for filename in os.listdir(src):
            src_file = os.path.join(src, filename)
            dst_file = os.path.join(dst, filename)
            if os.path.exists(dst_file):
                print(f"File {filename} already exists at the destination.")
                continue
            try:
                shutil.move(src_file, dst_file)
                print(f"Moved {filename}")
            except OSError as e:
                print(f"Error moving {filename}: {e}")
    except Exception as e:
        print(f"Error accessing source directory: {e}")

# Move files
move_files(source, destination)

# List files and directories in the destination path
print("After moving files:")
print(os.listdir(destination))

# Print the destination path
print("Destination path:", destination)
```

```python
import os

def count_files(directory):
    count = 0
    for root, dirs, files in os.walk(directory):
        count += len(files)
    return count

# Thay ƒë·ªïi ƒë∆∞·ªùng d·∫´n ƒë·∫øn th∆∞ m·ª•c b·∫°n mu·ªën ƒë·∫øm file
directory_path_1 = '/kaggle/working/yolov5/data_images/test'
directory_path_2 = '/kaggle/working/yolov5/data_images/train'
total_files_1 = count_files(directory_path_1)
total_files_2 = count_files(directory_path_2)
print(f'T·ªïng s·ªë file trong th∆∞ m·ª•c l√†: {total_files_1}')
print(f'T·ªïng s·ªë file trong th∆∞ m·ª•c l√†: {total_files_2}')
```
