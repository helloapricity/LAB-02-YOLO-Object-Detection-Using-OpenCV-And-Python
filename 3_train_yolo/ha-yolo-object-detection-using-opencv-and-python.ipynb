```python
import os
import shutil

# Source paths
source_train = '/kaggle/input/voc-dataset/data_images/data_images/train'
source_test = '/kaggle/input/voc-dataset/data_images/data_images/test'

# Destination paths
destination_base = '/kaggle/working/yolov5/data_images'
destination_train = os.path.join(destination_base, 'train')
destination_test = os.path.join(destination_base, 'test')

# Create destination directories if they don't exist
if not os.path.exists(destination_train):
    os.makedirs(destination_train)
if not os.path.exists(destination_test):
    os.makedirs(destination_test)

def copy_files(src, dst, num_pairs):
    txt_files = [f for f in os.listdir(src) if f.endswith('.txt')]
    jpg_files = [f for f in os.listdir(src) if f.endswith('.jpg')]
    
    moved_count = 0

    for txt_file in txt_files:
        jpg_file = txt_file.replace('.txt', '.jpg')
        if jpg_file in jpg_files:
            src_txt_path = os.path.join(src, txt_file)
            src_jpg_path = os.path.join(src, jpg_file)
            dst_txt_path = os.path.join(dst, txt_file)
            dst_jpg_path = os.path.join(dst, jpg_file)

            try:
                shutil.copy2(src_txt_path, dst_txt_path)
                shutil.copy2(src_jpg_path, dst_jpg_path)
                # print(f"Copied {txt_file} and {jpg_file}")
                moved_count += 1
            except OSError as e:
                print(f"Error copying {txt_file} and {jpg_file}: {e}")

            if moved_count >= num_pairs:
                break

    if moved_count < num_pairs:
        print(f"Only {moved_count} pairs were copied. {num_pairs} pairs were requested.")

# Copy 8000 pairs of train files
copy_files(source_train, destination_train, 8000)

# Copy 2000 pairs of test files
copy_files(source_test, destination_test, 2000)

# List files and directories in the destination path
print("After copying files:")
print("Train files:", len(os.listdir(destination_train)))
print("Test files:", len(os.listdir(destination_test)))

# Print the destination path
print("Destination path:", destination_base)
```

```python
import os

def count_files(directory):
    count = 0
    for root, dirs, files in os.walk(directory):
        count += len(files)
    return count

directory_path_test = '/kaggle/working/yolov5/data_images/test'
directory_path_train = '/kaggle/working/yolov5/data_images/train'
total_files_test = count_files(directory_path_test)
total_files_train = count_files(directory_path_train)
print(f'T·ªïng s·ªë file trong th∆∞ m·ª•c test l√†: {total_files_test}')
print(f'T·ªïng s·ªë file trong th∆∞ m·ª•c train l√†: {total_files_train}')
```
